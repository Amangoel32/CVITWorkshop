```
x = dict([(pow(10,r),r) for r in range(10)])
# Task 2.1: List all the keys inside the dict 
print(x.keys())
# Task 2.2: Remove key 100 from the dict 
del x[100]
# Task 2.3: Convert the above dict into a list of key, value tuple. Then sort the list based on mod of the key value with number 101
list = [(k, v) for k, v in x.items()] 
def com(x):
  return x[0]%101

list.sort(key=com)
print(list)
```

```
# Task 3: Calculate the inverse of a sqaure matrix using np.svd
# also verify the results X = np.ones((2,3,4))
import numpy as np
from numpy.linalg import inv

temp = np.random.rand(1000, 1000)
u, d, vh = np.linalg.svd(temp)
Ainv = vh.T@np.diag(1/d)@u.T
np.allclose(Ainv, inv(temp))
print(vh.T@np.diag(1/d)@u.T)
```

```
# Task 4: Create a sin plot which uses both plt.plot and plt.scatter. 
# Don't forget to add labels
%matplotlib inline
import matplotlib.pyplot as plt
x=np.linspace(-2*np.pi,2*np.pi,100)
y_sin=np.sin(x)

plt.plot(x,y_sin,label='Plot Graph')
plt.title('Plot Graph')
plt.show()
plt.scatter(x,y_sin,label='Scatter Graph')
plt.title('Scatter Graph')
plt.show()
```
